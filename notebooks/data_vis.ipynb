```python
import matplotlib.pyplot as plt
import rioxarray as rxr
import earthpy.plot as ep
from params import *
import os, glob, re, datetime
```

```python
#['CCHT', 'GSHT', 'BRHT', 'COHT', 'IVHT', 'AQHT', 'MGHT', 'ARHT', 'OKHT']
site_code = 'CCHT'
```

```python
def paths_to_datetimeindex(list):
    pattern = PATTERN_REGEX
    new_list = []
    for item in list:
        time = re.search(pattern, item).group(1)
        time = datetime.datetime.strptime(time, '%m-%d-%Y').date()
        new_list.append(time)
    new_list = sorted(new_list)    
    return new_list
```

```python
data_dir =  DOWNLOAD_DIR_ROOT + site_code + '/*.tif'
obsv = glob.glob(data_dir)
times = [i.strftime('%m-%d-%Y') for i in paths_to_datetimeindex(glob.glob(data_dir))]
times
```

```python
for i,ob in enumerate(obsv):
    ep.plot_rgb(rxr.open_rasterio(ob), rgb=[0, 1, 2], str_clip=0.9, stretch=True, title=ob)
```
